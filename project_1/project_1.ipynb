```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

## Import data from **global_data.csv** and calculate the average of 10 years

```python
os.getcwd
os.listdir(os.getcwd())
# data_world = pd.read_csv (r'C:\Users\Admin\Desktop\udacity\project_1\udacity\global_data.csv')
data_world = pd.read_csv ('global_data.csv')
#calculate average of 5 year
rolling_window=10
data_world['mean_of_10year'] = data_world['avg_temp'].rolling(rolling_window).mean()
print(data_world)
```

## *Draw the line chart 5 years moving average of Gobal*

```python

plt.figure(figsize=(20,10))
plt.plot(data_world['year'], data_world['mean_of_10year'], color='red', marker='o')
plt.title('Average temp Vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average rate', fontsize=14)
plt.grid(True)
plt.show()
```
